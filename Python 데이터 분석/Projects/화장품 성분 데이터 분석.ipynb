```python
# 화장품 데이터와 화장품 성분 데이터의 Ingredients 데이터를 맞추기 위해

# 1. 데이터 끝에 '.'이 있다면 제거
#    ex. 'Algae (Seaweed) Extract. Sea Salt.' 
#         -> 'Algae (Seaweed) Extract. Sea Salt'

# 2. 데이터 중간에 '. May Contain'을 포함하고 있다면
#    '. May Contain'을 포함한 이후의 데이터 삭제
#    ex. 'Algae (Seaweed) Extract. May Contain: Sea Salt, Fragrance' 
#         -> 'Algae (Seaweed) Extract'

# 3. ', '로 구분되어 있는 재료들 중 일부 재료 이름 수정
```

```python
# 1. 데이터 끝에 '.'이 있다면 제거
df_target['Ingredients'] = df_target['Ingredients'].str.rstrip('.')
df_target.head()
```

```python
# 2. 데이터 중간에 '. May Contain'을 포함하고 있다면
#    '. May Contain'을 포함한 이후의 데이터 삭제
df_target['Ingredients'] = df_target['Ingredients'].str.split('. May Contain').str[0]
df_target[df_target['Ingredients'].str.contains('. May Contain')]
```

```python
# 3. ', '로 구분되어 있는 재료들 중 일부 재료 이름 수정
replace_str_dict = {
    'Algae (Seaweed) Extract': 'Algae Extract',
    'Citrus Aurantifolia (Lime) Extract': 'Citrus Aurantifolia (Lime) Fruit Extract',
    'Eucalyptus Globulus (Eucalyptus) Leaf Oil': 'Eucalyptus Globulus Leaf Oil',
    'Galactomyces Ferment Filtrate (Pitera)': 'Galactomyces Ferment Filtrate',
    'Bacillus/Soybean/ Folic Acid Ferment Extract': 'Bacillus/Folic Acid/Soybean Ferment Extract',
    'Butyrospermum Parkii (Shea Butter)': 'Butyrospermum Parkii (Shea) Butter',
    'Sea Salt/Maris Sal/Sel Marin': 'Sea Salt',
    'Parfum/Fragrance': 'Fragrance|Perfume|Parfum',
    ', Fragrance': ', Fragrance|Perfume|Parfum',
    }

def replace_ingredients(ingredients):
    items = [ingredient.strip() for ingredient in ingredients.split(',')]
    items = [replace_str_dict.get(item, item) for item in items]
    return ', '.join(items)

df_target['Ingredients'] = df_target['Ingredients'].apply(replace_ingredients)
df_target.head()
```

```python
# Ingredients List 컬럼 생성

# 1. ', '로 구분되어 있는 Ingredients 컬럼의 데이터를
#    분리하여 list로 변환
#    ex. '  Algae (Seaweed) Extract,  Sea Salt  ' 
#         -> ['  Algae (Seaweed) Extract', ' Sea Salt  ']

# 2. 위 단계에서 변경한 list의 각 원소 앞뒤에 공백이 있다면 삭제
#    ex. ['  Algae (Seaweed) Extract', ' Sea Salt  '] 
#         -> ['Algae (Seaweed) Extract', 'Sea Salt']
```

```python
# 1. ', '로 구분되어 있는 Ingredients 컬럼의 데이터를
#    분리하여 list로 변환
df_target['Ingredients List'] = df_target['Ingredients'].str.split(', ')
df_target.head()
```

```python
# 2. 위 단계에서 변경한 list의 각 원소 앞뒤에 공백이 있다면 삭제
df_target['Ingredients List'] = df_target['Ingredients List'].apply(lambda x: [x_.strip() for x_ in x])
df_target.head()
```

```python
# 화장품 성분 데이터를 사용하여 화장품 데이터의
# 'Ingredient List'의 각 성분에 맞는 성분코드를 list에 저장하고
# 새로운 'Code List' 컬럼 생성
code_list = []

for ingredients_list in df_target['Ingredients List']:
    codes = []
    for ingredient in ingredients_list:
        ingredient_lower = ingredient.lower()

        matched = ingredients_df[ingredients_df['표준 영문명'].str.lower() == ingredient_lower]
        if len(matched):
            codes.append(matched['성분코드'].values[0])
        else:
            matched = ingredients_df[ingredients_df['구영문명'].str.lower() == ingredient_lower]
            if len(matched):
                codes.append(matched['성분코드'].values[0])
    code_list.append(codes)
df_target['Code List'] = code_list
df_target.head()
```

```python
# 'Code List'의 각 행에서 중복되지 않고,
# 모든 행에서 2번만 나온 코드를 오름차순으로 정렬하고,
# 상위 5개 데이터만 DataFrame으로 생성
unique_codes_by_row = []
for codes in df_target['Code List']:
    unique_codes_by_row.append(list(set(codes)))

unique_codes = []
for codes in unique_codes_by_row:
    unique_codes.extend(codes)

code_count_dict = {}
for code in unique_codes:
    if code in code_count_dict.keys():
        code_count_dict[code] += 1
    else:
        code_count_dict[code] = 1

codes_twice = [key for key, value in code_count_dict.items() if value == 2]
sorted_codes_twice_top_5 = sorted(codes_twice)[:5]

result_df = ingredients_df[ingredients_df['성분코드'].isin(sorted_codes_twice_top_5)]
result_df
```
